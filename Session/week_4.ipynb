```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df=pd.read_csv("./data (1).csv")
df.head(100)
df.info()
df.describe()
df.isnull().sum()
```

# Data Cleaning

```python

df.dropna(subset=["Calories"],inplace=True)
df.info()

## Duplication

print(df[df.duplicated(keep=False)])
print(df[df.duplicated(keep='first')])
print(df[df.duplicated(keep='last')])
```

# Cleaning

```python
df.drop_duplicates(inplace=True)
df.shape
```

```python
import numpy as np
def z_score(mean,x,sd):
    x=np.asarray(x,dtype=float)
    return (x-mean)/sd
```

# Outlier Detection

```python
print(df.describe())
z_score(64.522293,300,43.778959)
plt.scatter(df["Duration"],df["Calories"])
```

```python
df=df[df["Duration"]<192]
plt.scatter(df["Duration"],df["Calories"])
```
